```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt
import seaborn as se
```

## Importing the Dataset

```python
credit = pd.read_csv("data/tmdb_5000_credits.csv")
movies = pd.read_csv("data/tmdb_5000_movies.csv")
```

```python
credit.head()
```

```python
movies.head()
```

### Observation: 
    * movies and cradits have an common column Movie_id i.e. id.
    * Rename the column {"movie": "id"} and merge with movies in order to make whole complete dataframe.

```python
credit.rename(index=str,columns={"movie_id":"id"},inplace=True)
movies_merge = movies.merge(credit,on="id")
```

```python
movies_merge
```

```python
movies_merge.columns
```

### Observation:-
    * Some of the columns are not important role playing in recommandation system.
    * homepage: contaions the url of the movie.
    * Original_title, title_x and title_y are sme redendent data.
    * Status and production_countries are not required.

```python
movies_df = movies_merge.drop(["status","title_x","title_x","production_countries","status","homepage"],axis=1)
movies_df
```

```python
v=movies_df['vote_count']
R=movies_df['vote_average']
C=movies_df['vote_average'].mean()
m=movies_df['vote_count'].quantile(0.70)
```

```python
movies_df["Weighted-Rating"] = (((R*v)+(C*m))/(v+m))
movies_df
```

```python
col_data = movies_df[['original_title', 'vote_count', 'vote_average', 'Weighted-Rating', 'popularity']]
col_data.head(10)
```

```python
se.barplot(data=col_data.sort_values("vote_count",ascending=False).head(10),y="original_title",x="vote_count")
mlt.title("Vote Count")
```

```python
se.barplot(data=col_data.sort_values("vote_average",ascending=False).head(10),y="original_title",x="vote_average")
mlt.title("vote_average")
```

```python
se.barplot(data=col_data.sort_values("Weighted-Rating",ascending=False).head(10),y="original_title",x="Weighted-Rating")
mlt.title("Weighted-Rating")
```

```python
se.barplot(data=col_data.sort_values("popularity",ascending=False).head(10),y="original_title",x="popularity")
mlt.title("popularity")
```
