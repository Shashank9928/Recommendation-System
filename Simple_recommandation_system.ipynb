### Observation:

```python
credit.rename(index=str,columns={"movie_id":"id"},inplace=True)
movies_merge = movies.merge(credit,on="id")
```

```python
v=movies_df['vote_count']
R=movies_df['vote_average']
C=movies_df['vote_average'].mean()
m=movies_df['vote_count'].quantile(0.70)
```

```python
movies_df["Weighted-Rating"] = (((R*v)+(C*m))/(v+m))
movies_df
```

```python
col_data = movies_df[['original_title', 'vote_count', 'vote_average', 'Weighted-Rating', 'popularity']]
col_data.head(10)
```

### Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
col_data.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
col_data[["scaled_w_Rating","scaled_popularity"]] = mms.fit_transform(col_data[['Weighted-Rating','popularity']])
col_data
```

```python
col_data['score'] = col_data['scaled_w_Rating'] * 0.5 + col_data['scaled_popularity'] * 0.5
col_data.head()
```

### Here by using the "score" we reccomand the movies using collebrative base approch

```python
se.barplot(data=col_data.sort_values("score",ascending=False).head(10),y="original_title",x="score")
mlt.title("score")
```
